```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(city_data,ride_data, on="city")
# Display the data table for preview
combined_data = combined_data[['city','date','fare','ride_id','driver_count','type']]
```

```python
combined_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
grouped_by_city = combined_data.groupby(['city'])
cities = combined_data.drop_duplicates('city')

average_fare = grouped_by_city['fare'].mean()
number_drivers = grouped_by_city['driver_count'].mean()
riders = grouped_by_city['ride_id'].count()
city_type = cities.set_index('city')['type']

cleaned_data = pd.DataFrame({"Average Fair":average_fare, "Total Driver": number_drivers,
                             "Total Rider": riders,"City Type":city_type})

# Build the scatter plots for each city types
urban = cleaned_data.loc[cleaned_data["City Type"]=="Urban"]
suburban = cleaned_data.loc[cleaned_data["City Type"]=="Suburban"]
rural = cleaned_data.loc[cleaned_data["City Type"]=="Rural"]

u = plt.scatter(urban["Total Rider"],urban["Average Fair"], marker='o',s=urban["Total Rider"], label= "Urban", color='yellow',edgecolor = "black", alpha=0.5)
s = plt.scatter(suburban["Total Rider"],suburban["Average Fair"], marker='o',s=suburban["Total Rider"], label= "Surban", color='b', edgecolor = "black", alpha=0.5)
r = plt.scatter(rural["Total Rider"],rural["Average Fair"], marker='o',s=rural["Total Rider"], label= "Rural", color='g', edgecolor = "black", alpha=0.5)

# Incorporate the other graph properties
plt.xlabel("Total Number of Rider (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Date (2016)")
# Create a legend
plt.legend((u,s,r),("Urban", "Surban", "Rural"))
# Incorporate a text label regarding circle size

# Save Figure
plt.savefig("../Images/Pyber.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fare = grouped_by_city['fare'].sum()

total_fare = pd.DataFrame({"Total Fare":fare, "Total Driver": number_drivers,
                             "Total Rider": riders,"City Type":city_type})

urban_fare = total_fare.loc[total_fare['City Type'] == 'Urban']
suburban_fare = total_fare.loc[total_fare['City Type'] == 'Suburban']
rural_fare = total_fare.loc[total_fare['City Type'] == 'Rural']

urban_total = urban_fare["Total Fare"].sum()
suburban_total = suburban_fare["Total Fare"].sum()
rural_total = rural_fare["Total Fare"].sum()

# Build Pie Chart
label=["Urban", "Suburban", "Rural"]
sizes = [urban_total, suburban_total, rural_total]
color=["yellow", "blue", "green"]
explode = (0.1,0.1,0)

plt.pie(sizes, explode = explode, labels = label, autopct="%1.2f%%", startangle = 120)
plt.axis("equal")

plt.title=("% of Total Fare by City Type")
# Save Figure
```

```python
# Show Figure
plt.show()
```
